```python
import pandas as pd
import numpy as np

path = '/content/transferredPlayers-processed1780f.csv'
df = pd.read_csv(path)
print(df.shape)
df.head()
```

```python
#df.columns.values.tolist()
#df = df.drop(columns='Unnamed: 0')
#df = df.drop(columns='Unnamed: 0.1')
df.head()

df = df.drop(columns='Matches_NewT_FWD')
df = df.drop(columns='Matches_NewT_DEF')
df = df.drop(columns='Matches_NewT_MID')
```

```python
pd.set_option('display.max_rows', None)
df.isna().sum().sort_values(ascending=False)
```

```python
# dropping since many null values.
columns_to_drop = ['Mn/Sub_x', 'Mn/Start_x', 'G/SoT_y', 'G/SoT_x', 'Mn/Sub_y', 'npxG/Sh_y', 'Dist_y', 'G/Sh_y', 'SoT%_y', 'npxG/Sh_x', 'SoT%_x', 'G/Sh_x', 'Dist_x']
df = df.drop(columns=columns_to_drop)

pd.set_option('display.max_rows', None)
df.isna().sum().sort_values(ascending=False)
```

```python
# fill null with mean
df_filled = df.fillna(df.mean())
df_filled.isna().sum().sort_values(ascending=False)
```

```python
df_filled = df_filled.drop(columns='Unnamed: 0')
df_filled = df_filled.drop(columns='Unnamed: 0.1')
df_filled.head()
```

```python
# hot stop encoding for categorical fields
```

```python
df_filled = pd.get_dummies(df_filled, columns=['Comp_x'], prefix='Comp')
df_filled.head()
print(df_filled.columns.values.tolist())
```

```python
print(df.dtypes.to_string())
```

```python
df_filled = pd.get_dummies(df_filled, columns=['Pos_x'], prefix='Comp')
df_filled = pd.get_dummies(df_filled, columns=['Pos_y'], prefix='Comp')
df_filled = pd.get_dummies(df_filled, columns=['Comp_y'], prefix='Comp')


```

```python
df_filled.head()
```

```python
df_filled.to_csv('transferredPlayers-modelReady1780f.csv', index=True)
```

```python
#G+A_y
X = df_filled.drop(columns=["G+A_y"])
Y = df_filled["G+A_y"]
```

```python
# Pandas will allow us to create a dataframe of the data so it can be used and manipulated
import pandas as pd
# Regression Tree Algorithm
from sklearn.tree import DecisionTreeRegressor
# Split our data into a training and testing data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2, random_state=1)
```

```python
regression_tree = DecisionTreeRegressor(criterion = "squared_error")
```

```python
regression_tree.fit(X_train, Y_train)
```

```python
regression_tree.score(X_test, Y_test)
```

```python
from sklearn.metrics import mean_squared_error
prediction = regression_tree.predict(X_test)

print("$",(prediction - Y_test).abs().mean()*1000)
mse = mean_squared_error(Y_test, prediction)
print(mse)
```

```python
from sklearn.metrics import explained_variance_score

# Assuming you already have the prediction and target values
expl_variance = explained_variance_score(Y_test, prediction)
print("Explained Variance Score:", expl_variance)
```
